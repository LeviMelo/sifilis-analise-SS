```python
from ftputil import FTPHost

# FTP server and root directory
HOST   = "ftp.datasus.gov.br"
USER   = "anonymous"
PW     = ""

# Prefixes to match
PREFIXES = ("SIFC", "SIFG", "SIFA")

# Connect
host = FTPHost(HOST, USER, PW)

matched = []
for dirpath, dirnames, filenames in host.walk("/dissemin/publicos/SINAN"):
    for fn in filenames:
        if fn.startswith(PREFIXES):
            matched.append(host.path.join(dirpath, fn))

host.close()

print(f"Found {len(matched)} matching files:")
for path in matched:
    print(path)
```

```python
import ftputil

# FTP credentials and root path
FTP_HOST = "ftp.datasus.gov.br"
FTP_USER = "anonymous"
FTP_PASS = ""

# Prefixes of interest (upper-case expected in filenames)
PREFIXES = ("SIFA", "SIFC", "SIFG")
ROOT_PATH = "/dissemin"

# Connect to FTP server
host = ftputil.FTPHost(FTP_HOST, FTP_USER, FTP_PASS)

# List to store results
matched_files = []

# Recursive walk under DADOS
for dirpath, dirnames, filenames in host.walk(ROOT_PATH):
    for filename in filenames:
        # Match files starting with prefix AND ending in .dbc
        if filename.upper().startswith(PREFIXES) and filename.lower().endswith(".dbc"):
            full_path = host.path.join(dirpath, filename)
            matched_files.append(full_path)

host.close()

# Output results
print(f"âœ… Found {len(matched_files)} .dbc files starting with SIFA, SIFC, or SIFG:")
for path in matched_files:
    print(path)
```
